```python
# create a series and display the five names of IT company and their location. Change the index from 101 to 105.

import pandas as pd
a=pd.Series([['Google','Delhi'],
             ['Microsoft','Lucknow'],
             ['Flipkart','Bihar'],
             ['Dell','Ghaziabad'],
             ['HP','Patna']],index=[101,102,103,104,105])
print(a)
```

```python
# create a series and display the ten names of Movies and their rating in integer.

a = pd.Series({'Dhoom': 9, 'Dhoom2': 8, 'Dhoom3': 7, 'PK': 7, 'WAR': 6, 'Golmaal': 9, 'Golmaal Return': 8, 'Gadar': 7, 'Gadar 2': 7, 'Saaho': 6}, dtype=int)
print(a)
```

```python
# change the rating into float
a = pd.Series({'Dhoom': 9, 'Dhoom2': 8, 'Dhoom3': 7, 'PK': 7, 'WAR': 6, 'Golmaal': 9, 'Golmaal Return': 8, 'Gadar': 7, 'Gadar 2': 7, 'Saaho': 6}, dtype=float)
print(a)
```
