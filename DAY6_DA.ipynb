```python
import pandas as pd
df = pd.read_csv('mtcars2.csv')
df
```

```python
# 3 - unncessary data = S.No
df.drop(columns=['S.No']) #temporary data changed
```

```python
df.drop(columns=['S.No','hp','mpg'])
```

```python
df.drop(columns=['S.No'], inplace=True)# data is changed permanently
```

```python
# 2 - not for format = Unamed:
df.rename(columns={'Unnamed: 1': 'Name'}, inplace=True)
```
